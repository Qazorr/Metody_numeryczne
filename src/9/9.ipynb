```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.polynomial.polynomial as nppoly

x = np.arange(-7/8, 7/8+2/8, 2/8)
fx = np.array([1/(1+5*(val**2)) for val in x])

coofs = []
with open("9.data", "r") as inp:
    line = inp.readline()
    while(line):
        line = line.strip()
        c = line.split(" ")
        values = [float(x) for x in c]
        values = values[::-1]
        coofs.append(values)
        line = inp.readline()

polynomials = [nppoly.Polynomial(coof) for coof in coofs]

points = 1e4

colors = ["b-", "r-", "g-", "c-", "m-", "y-", "k-"]

plt.plot(x, fx, "rp")

for i in range(x.shape[0]-1):
    r = np.linspace(x[i], x[i+1], points)
    plt.plot(r, polynomials[i](r), colors[i])



plt.rcParams['figure.figsize'] = [10,5]

plt.show()
```
