```python
import matplotlib.pyplot as plt
import numpy as np

y1= np.loadtxt("GAUSS-SEIDEL_diff.data", delimiter = " ", unpack = True)
y2= np.loadtxt("GRADIENTS_diff.data", delimiter = " ", unpack = True)

fig = plt.figure(figsize = (12,6))
fig2 = plt.figure(figsize = (12,6))

p1 = fig.add_subplot(121)
p2 = fig.add_subplot(122)

p = fig2.add_subplot(111)

p1.plot(y1, color = "blue", label = "Gauss-Seidel")
#p1.set_xlim([y1[0], y1[-1]])
p1.set_xlabel("iterations")
p1.set_ylabel("||xk - xk-1||")
p1.set_title("Gauss-Seidel")

p2.plot(y2, color = "red", label="Gradients")
#p2.set_xlim([y2[0], y2[-1]])
p2.set_xlabel("iterations")
p2.set_ylabel("||xk - xk-1||")
p2.set_title("Gradients")

p.plot(y1, color = "blue", label = "Gauss-Seidel")
p.plot(y2, color = "red", label = "Gradients")
#p.set_xlim([-5, 50])
p.set_ylim([-2, 10])
p.set_xticks(np.arange(0, 50, 2))
p.set_xlabel("iterations")
p.set_ylabel("||xk - xk-1||")
p.set_title("Gauss-Seidel vs Gradients")
p.legend()

plt.show()
```
