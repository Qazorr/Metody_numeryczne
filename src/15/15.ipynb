```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.polynomial.polynomial as nppoly

x = []
fx = []

with open("15dane.data", "r") as data:
    line = data.readline()
    while(line):
        line = line.strip()
        parts = line.split(" ")
        x.append(float(parts[0]))
        fx.append(float(parts[1]))
        line = data.readline()


coofs = []
with open("15.data", "r") as inp:
    line = inp.readline()
    while(line):
        line = line.strip()
        c = line.split(" ")
        values = [float(x) for x in c]
        values = values[::-1]
        coofs.append(values)
        line = inp.readline()

polynomials = [nppoly.Polynomial(coof) for coof in coofs]

points = 1e4

for i in range(len(x)-1):
    r = np.linspace(x[i], x[i+1], points)
    plt.plot(r, polynomials[i](r), "r-")



plt.rcParams['figure.figsize'] = [10,5]

plt.show()
```
