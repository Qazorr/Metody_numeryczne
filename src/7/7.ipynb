```python
import numpy as np
import numpy.polynomial.polynomial as polyn
import matplotlib.pyplot as plt

coefficients = np.loadtxt("7.data")
coefficients = coefficients[::-1]

points_x = np.array([0.062500, 0.187500, 0.312500, 0.437500, 0.562500, 0.687500, 0.812500, 0.937500])
points_y = np.array([0.687959, 0.073443, -0.517558, -1.077264, -1.600455, -2.080815, -2.507266, -2.860307])
x = np.arange(-1, 1.1, 0.01)

polynomial = polyn.Polynomial(coefficients)

plt.plot(points_x, points_y, "rp", label = "points")
plt.plot(x, polynomial(x), "-", label = "polynomial")
plt.rcParams['figure.figsize'] = [10,5]
plt.legend()
plt.show()
```
