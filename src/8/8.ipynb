```python
import numpy as np
import numpy.polynomial.polynomial as polyn
import matplotlib.pyplot as plt

x = np.arange(-7/8, 7/8+2/8, 2/8)
fx = np.array([1/(1+5*(val**2)) for val in x])

coefficients = np.loadtxt("8.data")

#! Polynomial przyjmuje wspolczynniki w ascending order, a poly1d w descending
coefficients = coefficients[::-1]

polyn1 = polyn.Polynomial(coefficients)

r = np.arange(-1, 1+0.01, 0.01)

plt.plot(r, polyn1(r), "r-", label = "polynomial")
plt.plot(x, fx, "gp", label = "points")
plt.rcParams['figure.figsize'] = [10,5]
plt.legend()
plt.show()
```
